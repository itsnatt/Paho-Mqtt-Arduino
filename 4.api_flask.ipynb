```python
from flask import Flask, jsonify
import mysql.connector

app = Flask(__name__)

# Endpoint API untuk mendapatkan data terbaru dari tabel PreprocessingLog
@app.route('/api/data', methods=['GET'])
def get_latest_preprocessing_log():
    try:
        # Membuat koneksi
        conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="",
            database="IOT"
        )

        # Membuat kursor
        cursor = conn.cursor()

        # Menjalankan query untuk mendapatkan satu baris data terbaru
        cursor.execute("SELECT * FROM PreprocessingLog ORDER BY id DESC LIMIT 1")

        # Mengambil hasil query
        latest_log = cursor.fetchone()

        # Menutup kursor dan koneksi
        cursor.close()
        conn.close()

        # Mengonversi hasil ke format JSON
        if latest_log:
            result = {
                'id': latest_log[0],
                'node_id': latest_log[1],
                'id_node': latest_log[2],
                'kondisi_battery': latest_log[3],
                'status_hujan': latest_log[4],
                'akumulasi_kelembapan': latest_log[5],
                'perubahan_kemiringan': latest_log[6],
                'perubahan_lokasi': latest_log[7],
                'status_waspada': latest_log[8],
                'timestamp': str(latest_log[9])  # Convert timestamp ke string
            }
            return jsonify(result)
        else:
            return jsonify({'message': 'No preprocessing log found'})

    except Exception as e:
        # Handle exception jika terjadi kesalahan
        return jsonify({'error': str(e)})

if __name__ == '__main__':
    app.run(debug=False, port=1888)

```
